```python
new = {0:'No',1:'Yes'}
df['Survived'] = df.Survived.map(new)
```

```python
df
```

#### Exploratory Data Analysis

```python
ax = sns.countplot(data = df,x = 'Sex')
for x in ax.containers:
    ax.bar_label(x)
```

```python
titanic_sex = df.groupby(['Sex','Survived'],as_index = False)['Fare'].sum().sort_values('Fare',ascending = False)
sns.barplot(x = 'Sex',y ='Fare',hue = 'Survived',data = titanic_sex)
```

```python
ax = sns.countplot(data = df,x = 'SibSp')
for x in ax.containers:
    ax.bar_label(x)
```

```python
titanic_sib = df.groupby(['SibSp','Survived'],as_index = False)['Fare'].sum().sort_values('Fare',ascending = False)
sns.barplot(x = 'SibSp',y ='Fare',hue= 'Survived',data = titanic_sib)
```

```python
ax = sns.countplot(data = df,x = 'Parch')
for x in ax.containers:
    ax.bar_label(x)
```

```python
titanic_Parch = df.groupby(['Parch','Survived'],as_index = False)['Fare'].sum().sort_values('Fare',ascending = False)
sns.barplot(x = 'Parch',y ='Fare',hue = 'Survived',data = titanic_Parch)
```

```python
df.columns
```

```python
ax = sns.countplot(data = df,x = 'Pclass')
for x in ax.containers:
    ax.bar_label(x)
```

```python
titanic_Pclass = df.groupby(['Pclass','Survived'],as_index = False)['Fare'].sum().sort_values('Fare',ascending = False)
sns.barplot(x = 'Pclass',y ='Fare',hue = 'Survived',data = titanic_Pclass)
```
